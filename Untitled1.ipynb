```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rawData = pd.read_csv('production-demi-horaire-agregee-par-region.csv', delimiter=';')
print(rawData.head())
rawData.describe()
```

```python
# separate National data
nationalData = rawData[rawData['region'] == 'National']
prodData = rawData[rawData['region'] != 'National']
print(nationalData.describe())
prodData.describe()
```

```python
# clean national data
cleanNationalData = nationalData.copy()
cleanNationalData = cleanNationalData.drop(columns=['region', 'grd'])
cleanNationalData['date'] = pd.to_datetime(cleanNationalData['horodate'])
cleanNationalData = cleanNationalData.drop(columns=['horodate'])
cleanNationalData['day'] = cleanNationalData['date'].dt.date
cleanNationalData['time'] = cleanNationalData['date'].dt.time
```

```python
# group national data by day
groupedNationalDataByDay = cleanNationalData.groupby(['day'])['energie_injectee'].mean()

# plot national data
groupedNationalDataByDay.plot()
```

```python
# group national data by time
groupedNationalDataByTime = cleanNationalData.groupby(['time'])['energie_injectee'].mean()

# plot national data
groupedNationalDataByTime.plot()
```

```python
# Create a new column 'day_of_year' that represents the day of the year
cleanNationalData['day_of_year'] = cleanNationalData['date'].dt.dayofyear

# Group by 'day_of_year' and calculate the mean of 'energie_injectee'
groupedNationalDataByDayOfYear = cleanNationalData.groupby('day_of_year')['energie_injectee'].mean()

fig, ax = plt.subplots(figsize=(10, 5))

# Dfinir les couleurs de chaque saison
seasons = {
    'Spring': (80, 172, 'green'),
    'Summer': (173, 264, 'red'),
    'Autumn': (265, 355, 'orange'),
    'Winter1': (1, 79, 'blue'),
    'Winter2': (356, 365, 'blue')
}

# Ajouter des zones colores pour chaque saison
for season, (start, end, color) in seasons.items():
    ax.axvspan(start, end, color=color, alpha=0.3, label=season)

# Tracer les donnes
groupedNationalDataByDayOfYear.plot(ax=ax, color='black')

# Lgende
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))  # liminer les doublons dans la lgende
plt.legend(by_label.values(), by_label.keys())

# Mettre des titres et des labels
plt.title('Mean Injected Energy by Day of Year with Seasonal Backgrounds')
plt.xlabel('Day of Year')
plt.ylabel('Mean Injected Energy (Unit)')

# Afficher le graphique
plt.show()
```

```python
# clean prod data
cleanProdData = prodData.copy()
cleanProdData = cleanProdData.drop(columns=['region', 'grd'])
cleanProdData['date'] = pd.to_datetime(cleanProdData['horodate'])
cleanProdData = cleanProdData.drop(columns=['horodate'])
cleanProdData['day'] = cleanProdData['date'].dt.date
cleanProdData['time'] = cleanProdData['date'].dt.time
```

```python
# visualize difference between winter data over the years
winterData = cleanProdData[(cleanProdData['date'].dt.month == 1) | (cleanProdData['date'].dt.month == 2) | (cleanProdData['date'].dt.month == 12)]
winterData = winterData.groupby(['day'])['energie_injectee'].mean()

plt.bar(winterData.index, winterData)
plt.xlabel('Day')
plt.ylabel('Average Injected Energy')
plt.title('Average Injected Energy per Day in Winter')
plt.show()
```

```python
# Create a new column 'winter' that represents the winter season
cleanProdData['winter'] = cleanProdData['date'].apply(lambda x: x.year if x.month > 2 else x.year - 1)

# Filter winter data
winterData = cleanProdData[(cleanProdData['date'].dt.month == 1) | (cleanProdData['date'].dt.month == 2) | (cleanProdData['date'].dt.month == 12)]

# Group by winter and calculate mean and max energy
meanMaxData = winterData.groupby(['winter'])['energie_injectee'].agg(['mean', 'max'])

# Print mean and max consumption for each winter
print(meanMaxData)
```

```python
# Plot mean and max consumption for each winter
plt.figure(figsize=(10, 6))
plt.plot(meanMaxData.index, meanMaxData['mean'], marker='o', label='Mean')
plt.plot(meanMaxData.index, meanMaxData['max'], marker='o', label='Max')

# Adjust xticks to only show full years
plt.xticks(np.arange(meanMaxData.index.min(), meanMaxData.index.max()+1, 1), rotation=45)

plt.xlabel('Winter')
plt.ylabel('Injected Energy')
plt.title('Mean and Max Injected Energy per Winter')
plt.legend()
plt.show()
```
